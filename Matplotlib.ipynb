#### Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

```python
Matplotlib is a popular Python library for data visualization. It provides a wide variety of static, animated,
and interactive plots for exploring and presenting data in a visual form. Matplotlib is often used for creating
publication-quality charts, graphs, and other visualizations in scientific computing, data analysis, and machine learning.

The pyplot module is a collection of functions in Matplotlib that allows you to create various types of plots quickly.
Here are five common types of plots that can be created using the pyplot module:
```

```python
## line plot

import matplotlib.pyplot as plt
import numpy as np
x = np.random.rand(50)
y = np.random.rand(50)
plt.plot(x,y)
```

```python
## scatter plot

p = np.random.rand(50)

plt.scatter(x,y, c = p )
plt.xlabel("x-axis")
plt.ylabel('y-axis')
plt.title("Distribution between x and y axis")
plt.show()
```

```python
## HIstogram

plt.hist(x)
plt.show()
```

```python
## Bar chart

plt.bar(x,y)
plt.show()
```

```python
## pie chart

import matplotlib.pyplot as plt

labels = ['Category A', 'Category B', 'Category C']
sizes = [30, 45, 25]

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=['lightcoral', 'lightblue', 'lightgreen'])
plt.title('Pie Chart')
plt.show()
```

#### Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

```python
scatter plot is a type of data visualization that displays individual data points on a two-dimensional graph.
Each point represents the values of two variables, one plotted along the x-axis and the other along the y-axis. 
Scatter plots are useful for visualizing the relationship between two continuous variables and identifying patterns
or trends in the data.
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
np.random.seed(3)
```

```python
x = np.random.normal(0,2,50)   # numpy.random.normal(loc, scale, size)
y = np.random.normal(0,2,len(x))
```

```python
## scatter plot

g = np.random.rand(50)
plt.scatter(x,y, c = g, marker = '*')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title("the distribution of x and y")
plt.show()
```

#### Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

```python
The subplot() function in Matplotlib is used to create multiple plots within the same figure. It allows you to arrange
multiple subplots in a grid, specifying the number of rows and columns in the grid. This is useful  when you want to
compare different datasets or visualize multiple aspects of your data in a single figure
```

```python
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])
```

```python
## line 1

x = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

plt.subplot(2,2,1)  #  # 2 rows, 2 columns, 1st subplot
plt.plot(x,y1, color = 'r')
plt.title("plot 1")


## line 2

x = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

plt.subplot(2,2,2)
plt.scatter(x,y2)
plt.title('plot 2')


## line 3

x = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

plt.subplot(2,2,3)
plt.scatter(x,y3)
plt.title('plot 3')



## line 4

x = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

plt.subplot(2,2,4)
plt.scatter(x,y4)
plt.title('plot 4')

plt.tight_layout()

plt.show()
```

#### Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot. import numpy as np

```python
A bar plot is a graphical representation of categorical data with rectangular bars. The length of each bar corresponds
to the value of a particular category, making it easy to compare the quantities across different categories. 
Bar plots are commonly used to visualize the distribution or comparison of discrete categories.
```

```python
You can use the 'bar()' function to create vertical bar plots and the 'barh()' function to create horizontal bar plots.
```

```python
## bar plot
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])


plt.bar(company,profit, color = 'g')
plt.title("The Company's profit")
plt.xlabel('Company')
plt.ylabel('profit')
plt.show()
```

```python
## horizontal barplot

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])


plt.barh(company,profit, color = 'b')
plt.title("The Company's profit")
plt.xlabel('Company')
plt.ylabel('profit')
plt.show()
```

#### Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

```python
A box plot (box-and-whisker plot) is a statistical visualization that provides a summary of the distribution of a dataset. 
It displays key summary statistics, such as the median, quartiles, and potential outliers. The box in the plot represents 
the interquartile range (IQR) between the first quartile (Q1) and the third quartile (Q3), while the line inside the box 
represents the median. Whiskers extend from the box to the minimum and maximum values within a specified range, and points 
beyond the whiskers are considered outliers.

Box plots are particularly useful for comparing the distribution of multiple datasets or identifying the presence of outliers
in a single dataset.
```

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
plt.figure(figsize = (8,6))
plt.boxplot(box1, labels=['Dataset 1'])
plt.xlabel('Datasets')
plt.ylabel('Values')
plt.title('Box Plot of Datasets')
plt.show()
```

```python
plt.figure(figsize = (8,6))
plt.boxplot(box2, labels=['Dataset 2'])
plt.xlabel('Datasets')
plt.ylabel('Values')
plt.title('Box Plot of Datasets')
plt.show()
```
